## Assignment 10 - A43 CNN Architecture Breakdown
Conv2D Layers extract low-level (edges, textures) and then higher-level (digits, shapes) features.

ReLU Activation removes negative values, introducing non-linearity.

MaxPooling reduces feature map size and helps avoid overfitting.

Flatten bridges convolutional layers to fully connected layers (Dense).

Dense + Softmax makes predictions by classifying into 10 digit classes (0–9).

```python
import tensorflow as tf
import warnings
warnings.filterwarnings('ignore')

# Load the MNIST dataset - Loads MNIST digits (28×28 grayscale images).
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize pixel values to [0, 1]
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Add a channel dimension to the images makes it (28, 28, 1)
x_train = x_train[..., tf.newaxis]
x_test = x_test[..., tf.newaxis]

#Set up the Layers of the network
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)), # 32 filters of size 3×3 + ReLU
    tf.keras.layers.MaxPooling2D((2, 2)),# Downsamples the feature map
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'), # It applies a set of learnable filters (kernels) that convolve across the input image to detect spatial features like edges or patterns.
    tf.keras.layers.MaxPooling2D((2, 2)), # It reduces the spatial dimensions, making computation efficient and helping to retain only the most important features.
    tf.keras.layers.Flatten(),# Converts 2D data to 1D. CNN outputs a 3D feature map; Flatten converts it into a 1D vector so it can be fed into the fully connected (dense) layers.
    tf.keras.layers.Dense(10, activation="softmax")]) # Output layer for 10 digit classes. Softmax outputs probabilities for multi-class classification, ensuring all outputs sum to 1 and helping to pick the most likely digit.

#Compile the model
model.compile(optimizer="adam",
              loss ="sparse_categorical_crossentropy",# Because the labels (y_train, y_test) are integers (0–9) and not one-hot encoded. Sparse version handles this directly.
              metrics=["accuracy"])

#Train the model on the MNIST dataset
model.fit(x_train, y_train, epochs=3, validation_data=(x_test, y_test))

#Evaluate the model on the MNIST test data
test_loss, test_acc = model.evaluate(x_test, y_test)
print('Test accuracy: ', test_acc)
```

```python
# Plot confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
y_pred = model.predict(x_test).argmax(axis=1)# The model outputs probability vectors. argmax picks the class with the highest probability as the predicted digit.
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title(f'Test Accuracy: {test_acc:.2f}')
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Load and normalize data
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
x_train = x_train[..., tf.newaxis]
x_test = x_test[..., tf.newaxis]

# Modified CNN model
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.BatchNormalization(), # It normalizes the inputs of a layer to stabilize and speed up training by reducing internal covariate shift.
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.25), # Dropout randomly deactivates a fraction of neurons during training, which prevents overfitting and helps the model generalize better.

    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.25),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(10, activation='softmax') 
])

# Compile and train
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(x_train, y_train, epochs=3, validation_data=(x_test, y_test))

# Evaluate model
test_loss, test_acc = model.evaluate(x_test, y_test)
print('Test Accuracy:', test_acc)

# Plot confusion matrix
y_pred = model.predict(x_test).argmax(axis=1)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title(f'Test Accuracy: {test_acc:.2f}')
plt.show()
```
