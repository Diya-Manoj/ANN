```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear(x):
    return x

x = np.linspace(-10,10,100)
plt.plot(x, linear(x), label = 'Linear Activation', color = 'r')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Linear Activation Function')
plt.grid()
plt.legend()
plt.show()
```

```python
def sigmoid(x):
    return 1/1 + np.exp(-x)

x = np.linspace(-10,10,100)
plt.plot(x, 1/(1 + np.exp(-x)), label = 'Sigmoid Activation', color = 'b')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Sigmoid Activation Function')
plt.grid()
plt.legend()
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)

x = np.linspace(-10,10,100)
plt.plot(x, np.tanh(x), label = 'Tanh Activation', color = 'g')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Tanh Activation Function')
plt.grid()
plt.legend()
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-10,10,100)
plt.plot(x, np.maximum(0, x), label = 'ReLU Activation', color = 'm')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('ReLU Activation Function')
plt.grid()
plt.legend()
plt.show()
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum()

x = np.linspace(-10, 10, 100)  # Using a reasonable range
y = softmax(x)
plt.plot(x, y, label = 'SoftMax Activation', color = 'c')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('SoftMax Activation Function')
plt.grid()
plt.legend()
plt.show()
```

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, x * alpha)

x = np.linspace(-10, 10, 100)
plt.plot(x, leaky_relu(x), label='Leaky ReLU', color='orange')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Leaky ReLU Activation')
plt.grid()
plt.legend()
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def linear(x):
    return x

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # For numerical stability
    return exp_x / exp_x.sum()

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, x * alpha)

x = np.linspace(-5, 5, 100)

plt.figure(figsize=(12, 8))

# Sigmoid
plt.subplot(2, 3, 1)
plt.plot(x, sigmoid(x), label='Sigmoid', color='b')
plt.title('Sigmoid Activation')
plt.grid()
plt.legend()

# ReLU
plt.subplot(2, 3, 2)
plt.plot(x, relu(x), label='ReLU', color='r')
plt.title('ReLU Activation')
plt.grid()
plt.legend()

# Tanh
plt.subplot(2, 3, 3)
plt.plot(x, tanh(x), label='Tanh', color='g')
plt.title('Tanh Activation')
plt.grid()
plt.legend()

# Linear
plt.subplot(2, 3, 4)
plt.plot(x, linear(x), label='Linear', color='m')
plt.title('Linear Activation')
plt.grid()
plt.legend()

# SoftMax
plt.subplot(2, 3, 5)
plt.plot(x, softmax(x), label='SoftMax', color='c')
plt.title('SoftMax Activation')
plt.grid()
plt.legend()

# Leak Relu
plt.subplot(2, 3, 6)
plt.plot(x, leaky_relu(x), label='Leaky Relu', color='orange')
plt.title('Leaky ReLU Activation')
plt.grid()
plt.legend()


plt.tight_layout()
plt.show()
```

```python
def choose_activation():
    import numpy as np
    import matplotlib.pyplot as plt

    def sigmoid(x): return 1 / (1 + np.exp(-x))
    def tanh(x): return np.tanh(x)
    def relu(x): return np.maximum(0, x)
    def leaky_relu(x, alpha=0.01): return np.where(x > 0, x, x * alpha)
    def linear(x): return x
    def softmax(x): 
        exp_x = np.exp(x - np.max(x))
        return exp_x / exp_x.sum()

    activations = {
        'sigmoid': sigmoid,
        'tanh': tanh,
        'relu': relu,
        'leaky_relu': leaky_relu,
        'linear': linear,
        'softmax': softmax
    }

    print("Available activations:", list(activations.keys()))
    act = input("Enter activation function name: ").strip().lower()
    start = float(input("Enter range start: "))
    end = float(input("Enter range end: "))

    x = np.linspace(start, end, 100)
    y = activations[act](x) if act != "leaky_relu" else activations[act](x, 0.01)

    plt.plot(x, y, label=f'{act} activation', color='purple')
    plt.title(f'{act.capitalize()} Activation')
    plt.grid()
    plt.legend()
    plt.show()

choose_activation()
```

```python
x = np.linspace(-5,5,25)

plt.plot(x, sigmoid(x), label='Sigmoid (0 to 1)', linestyle='--')
plt.plot(x, tanh(x), label='Tanh (-1 to 1)', linestyle='-.')
plt.plot(x, relu(x), label='ReLU (0 to infinity)', linestyle='-')
plt.plot(x, leaky_relu(x), label='Leaky ReLU (-infinity to infinity)', linestyle=':')
plt.plot(x, linear(x), label='Linear', linestyle='dotted')
plt.plot(x, softmax(x), label='Softmax (0 to 1)')

plt.title('Overlapping Activation Functions')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid()
plt.show()
```
